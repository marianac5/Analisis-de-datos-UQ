Leer los modulos necesarios para el calculo de los indicadores de Mercado Laboral:
Caracteristicas generales, seguridad social en salud y educacion
Ocupados

```python
#Caracteristicas generales, seguridad social en salud y educacion

# Links a los conjuntos de datos
url_cgsse = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/cgsse.CSV'
url_ocupados = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Ocupados.CSV'
url_ft = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Fuerza%20de%20trabajo.CSV'
url_desocupados =  'https://raw.githubusercontent.com/niconomist98/DataAnalyticsUQ/refs/heads/main/Datos/GEIH2025/No%20ocupados.CSV'

# CGSSE
df_cgsse = pd.read_csv(url_cgsse, encoding='latin-1', sep=';', 
                       low_memory=False)

#Ocupados
df_ocupados=pd.read_csv(url_ocupados,encoding='latin-1',sep=';',
                        low_memory=False)

#Fuerza de trabajo
df_ft=pd.read_csv(url_ft,
                  encoding='latin-1',sep=';',low_memory=False)
#No ocupados
df_desocupados=pd.read_csv(url_desocupados,encoding='latin-1',
                           sep=';',low_memory=False)
```

Calcular población total

```python
PT =  df_cgsse['FEX_C18'].sum()
print(f' La población total de referencia para Colombia basado en el modulo cgsse de la GEIH y el factor de expansión es {PT}')
```

Calcular población total : desagregación hombres y mujeres

```python
#P3271 : Cuál fue su sexo al nacer? Masculino: 1 Femenino: 2

PH = df_cgsse[df_cgsse['P3271']==1]['FEX_C18'].sum()
PM = df_cgsse[df_cgsse['P3271']==2]['FEX_C18'].sum()
print(f' Hombres : {PH}')
print(f' Mujeres:  {PM}')
```

```python
PH/PT*100
```

```python
PM/PT*100
```

```python
#CALCULAR POBLACION TOTAL DEL QUINDIO
#DPTO : departamento , Quindio es el 63

PTQ = df_cgsse[df_cgsse['DPTO']==63]['FEX_C18'].sum()
PTQ
```

```python
#Calculando cantidad de hombres y de mujeres
PHQ = df_cgsse[(df_cgsse['DPTO']==63) & (df_cgsse['P3271']==1)]['FEX_C18'].sum()
PMQ = df_cgsse[(df_cgsse['DPTO']==63) & (df_cgsse['P3271']==2)]['FEX_C18'].sum()
print(f' Hombres del Quindio : {PHQ}')
print(f' Mujeres del Quindio:  {PMQ}')
```

```python
#Porcentaje de hombres en el Quindio
PHQ/PTQ*100
```

```python
#Porcentaje de mujeres en el Quindio

PMQ/PTQ*100
```
